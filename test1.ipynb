**Pandas**
1. Pandas is python library which can be use for data analysis.
2. Intially Pandas are called as panel data
3. Pandas is open Source python library which can be use for data manipulation.
4. using pandas we can create data series as well as data frame.
5. Data series represent data in single dimension array format
6. Data frame represent data in two dimension array format

**Advantage :**

1. easy to learn
2. very powerfull for data analysis
3. save time for developer to manual work

**Disadvantage**

1. USe more memory for large data processing
2. not ideal for real time big data analysis.

**Exercise 12. Combine data from multiple data frame**

```python
data1 = {"Sno":[1,2],"Name":["Afzal","Arshad"],
         "Age":[32,30] , "Address":["Allahabad","Basti"]}
df1 = pd.DataFrame(data1)
df1
```

```python
data2 = [
    [3,"Amjad",27,"New Delhi"],
    [4,"Asjad",23,"Chandigarh"],
]
df2 = pd.DataFrame(data2,columns=["Sno","Name","Age","Address"])
df2
```

```python
df3= pd.concat([df1,df2])
df3
```
