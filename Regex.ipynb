                                    **#Regular Expression :**

 1. Regular Expression is declarative mechanism which is used to represent group of string according to particular pattern or format,

 2. Regular Expression can be used for data filter and validation

Regular expression module contain multiple function.

compile() : it is used to compile pattern into regex object
finditer() : it is used to find number of iteration of matching object.

Start() : starting index of matching object
End() : end+1 index of matching object
group() : group of matching

```python
import re
n = input("Enter a string: ")
p = re.compile("ab")
m = p.finditer(n)
count = 0
for i in m:
    count = count + 1
    print(i.start(), i.end(),i.group())
print("number of matches: ", count)
```


                            **#charactor classes <br>**
we can use charactor classes to search group of strings.

[abc] - either a or b or c  <br>
[^abc] - except a or b or c  <br>
[a-z] - a to z lower case    <br>
[A-Z] - A to Z upper case    <br>
[a-zA-Z] - a to z lower and capital letter  <br>
[a-zA-Z0-9] - alphanumeric charactor   <br>
[^a-zA-Z0-9]  - except alphanumeric (special charactor)   <br>

#\s - space charactor <br>
#\S - except Space charactor  <br>
#\d - digit   <br>
#\D - except digit  <br>
#\w - any charactor  <br>
#\W - except charactor  <br>
#. - any charactor including special charactor  <br>

#we can use quantifiers to specify the number of occurrence to matches  <br>

#a -- exact only a   <br>
#a+ -- At least one a  <br>
#a* -- any number of a's including zero number  <br>
#a? -- almost one a either zero number or one number   <br>
#a{m} -- exactly m number of a's  <br>
#a{m,n} -- minimum m number of a's and maximum n umber of a's  <br>

```python
import re

n = input("Enter a string : ")

m = re.finditer("[abc]",n)
count = 0
for i in m:
    count = count + 1
    print(i.start(), i.end(),i.group())
print("number of matches: ", count)
```

^ Symbol - for starting of target string  <br>
$ symbol - for ending of target string

```python
s = input("Enter a string : ")
m = re.search("^hello",s,re.IGNORECASE)
if m!=None:
    print("target string start with hello")
else:
    print("target string not start with hello")
```

```python
s = input("Enter a string : ")
m = re.search("World$",s,re.IGNORECASE)
if m!=None:
    print("target string end with world")
else:
    print("target string not end with world")
```

**Regular Expression for 10 digit mobile number validation**<br>

**1. Mobile number should 10 digit**<br>
**2. Mobile number should start with 6, 7, 8 or 9**<br>


```python
mobile_number = input("Enter a your string : ")
m = re.fullmatch("[6-9]{1}\\d{9}",mobile_number)
if m!=None:
    print("mobile number is valid")
else:
    print("mobile number not valid")
print(m)
```
